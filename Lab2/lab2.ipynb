```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  
df = sns.load_dataset('diamonds')

#   
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# 1. Scatter Plot: '   (carat)   (price)
plt.figure()
sns.scatterplot(data=df.sample(2000), x="carat", y="price", hue="cut", alpha=0.6)
plt.title("1.  vs  (    )")
plt.show()
# 

# 2. Histogram:   
plt.figure()
sns.histplot(df['price'], bins=30, kde=True, color="teal")
plt.title("2.    ")
plt.show()

# 3. Box Plot:      (cut)
plt.figure()
sns.boxplot(data=df, x="cut", y="price", palette="Set2")
plt.title("3.     ")
plt.show()
# 

# 4. Violin Plot:      (color)
plt.figure()
sns.violinplot(data=df, x="color", y="carat", split=True, inner="quart")
plt.title("4.   (carat)    ")
plt.show()

# 5. Bar Plot:      (clarity)
plt.figure()
sns.barplot(data=df, x="clarity", y="price", estimator=np.mean, palette="viridis")
plt.title("5.     ")
plt.show()

# 6. Heatmap:     (x, y, z, depth, table)
plt.figure()
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("6.     ")
plt.show()
# 

# 7. Joint Plot:     
g = sns.jointplot(data=df.sample(2000), x="carat", y="price", kind="hex", color="#4CB391")
g.fig.suptitle("7.  ' / (Hexbin)", y=1.02)
plt.show()

# 8. Point Plot:       
plt.figure()
sns.pointplot(data=df, x="color", y="price", hue="cut", dodge=True, markers=["o", "s", "D", "x", "^"])
plt.title("8.   :  vs ")
plt.show(),
```
